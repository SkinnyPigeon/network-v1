```python
import pandas as pd
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 2
xs = [i * m for i in range(50)]
c = 50
```

```python
ms = [m for _ in range(50)]
cs = [c for _ in range(50)]
```

```python
ys = [m * x + c for x in xs]
```

```python
plt.plot(xs, ys)
```

```python
data = {
    'ms': ms,
    'xs': xs,
    'cs': cs,
    'ys': ys
}

df = pd.DataFrame(data=data)
```

```python
df.head(5)
```

```python
X_train = df[['ms', 'xs', 'cs']]
y_train = df['ys']
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=3, input_shape=[3]))
model.add(tf.keras.layers.Dense(units=1, input_shape=[3]))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs=500)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
model.get_weights()
```

```python
x = 10

tdata = {
    'ms': [2],
    'xs': [x],
    'cs': [50]
}
test = pd.DataFrame(tdata)
test
```

```python
model.predict(test)
```

```python
def predict(input):
    keys = input.keys()
    output = 0
    for layer in model.layers:
#         print(layer.weights)
#         print("")
        for i in range(len(keys)):
            value = input[keys[i]][0]
            print(layer.weights[0][i])
            value *= layer.weights[0][i]
            output += value
        output += layer.weights[1]
    return output
```
